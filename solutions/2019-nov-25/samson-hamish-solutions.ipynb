````python
from collections import defaultdict

def slice_numbers(lines):
    d = defaultdict(list)
    ll = max(len(l) for l in lines)
    for line in lines:
        for i in range(0, ll, 3):
            snippet = line[i:i+3]
            d[i//3].append(snippet)
    return d

def get_slices(lines):
    l = []
    slices = slice_numbers(lines)
    for i in range(10):
        l.append(''.join(slices[i]))
    return l

one_to_nine = '''
 _     _  _     _  _  _  _  _ 
| |  | _| _||_||_ |_   ||_||_|
|_|  ||_  _|  | _||_|  ||_| _|
'''

base_slices = get_slices(one_to_nine.split('\n'))

base_dict = {''.join(v): k for k, v in slice_numbers(one_to_nine.split('\n')).items()}

for k, v in base_dict.items():
    print(f'"{k}" => {v}')

def get_account_number(slices):
    return ''.join(str(base_dict.get(s, '?')) for s in slices[:9])

def checksum(num):
    num = str(num)
    a = 0
    for i in range(9, 0, -1):
        a += int(num[i-1]) * (10 - i)
    return a % 11 == 0


def get_result(ac):
    if '?' in ac:
        return ac, 'ILL'
    elif not checksum(ac):
        return ac, 'ERR'
    else:
        return ac, 


test_cases = '''
 _  _  _  _  _  _  _  _  _ 
| || || || || || || || || |
|_||_||_||_||_||_||_||_||_|

                           
  |  |  |  |  |  |  |  |  |
  |  |  |  |  |  |  |  |  |

 _  _  _  _  _  _  _  _  _ 
 _| _| _| _| _| _| _| _| _|
|_ |_ |_ |_ |_ |_ |_ |_ |_ 

 _  _  _  _  _  _  _  _  _ 
 _| _| _| _| _| _| _| _| _|
 _| _| _| _| _| _| _| _| _|

                           
|_||_||_||_||_||_||_||_||_|
  |  |  |  |  |  |  |  |  |

 _  _  _  _  _  _  _  _  _ 
|_ |_ |_ |_ |_ |_ |_ |_ |_ 
 _| _| _| _| _| _| _| _| _|

 _  _  _  _  _  _  _  _  _ 
|_ |_ |_ |_ |_ |_ |_ |_ |_ 
|_||_||_||_||_||_||_||_||_|

 _  _  _  _  _  _  _  _  _ 
  |  |  |  |  |  |  |  |  |
  |  |  |  |  |  |  |  |  |

 _  _  _  _  _  _  _  _  _ 
|_||_||_||_||_||_||_||_||_|
|_||_||_||_||_||_||_||_||_|

 _  _  _  _  _  _  _  _  _ 
|_||_||_||_||_||_||_||_||_|
 _| _| _| _| _| _| _| _| _|

    _  _     _  _  _  _  _ 
  | _| _||_||_ |_   ||_||_|
  ||_  _|  | _||_|  ||_| _|

 _  _  _  _  _  _  _  _    
| || || || || || || ||_   |
|_||_||_||_||_||_||_| _|  |

    _  _  _  _  _  _     _ 
|_||_|| || ||_   |  |  | _ 
  | _||_||_||_|  |  |  | _|

    _  _     _  _  _  _  _ 
  | _| _||_| _ |_   ||_||_|
  ||_  _|  | _||_|  ||_| _ 
'''
    
def handle(input_str):
    for block in input_str.split('\n\n'):
        print(block)
        slices = get_slices(block.split('\n'))
        ac = get_account_number(slices)
        #print(list(enumerate(slices)))
        print(f'=> {" ".join(get_result(ac))}')
        
handle(test_cases)
````

## Kata 4 - Fizz Buzz Plus

````python
def fizzbuzz(start=1, end=100):
    for n in range(start, end + 1):
        if (n % 3 == 0 or '3' in str(n)) and (n % 5 == 0 and '5' in str(n)):
            yield 'FizzBuzz'
        elif n % 3 == 0 or '3' in str(n):
            yield 'Fizz'
        elif n % 5 == 0 or '5' in str(n):
            yield 'Buzz'
        else:
            yield n
            
list(fizzbuzz())
````
